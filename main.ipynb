##### Store Dataframe by reading from  2023_åk9_np.xlsx and add them to a list

```python
df_en = pd.read_excel("riket2023_åk9_np.xlsx",header= 8,sheet_name="Engelska")
df_se = pd.read_excel("riket2023_åk9_np.xlsx",header= 8,sheet_name="Svenska")
df_math = pd.read_excel("riket2023_åk9_np.xlsx",header= 8,sheet_name="Matematik")
df_sfi = pd.read_excel("riket2023_åk9_np.xlsx",header= 8,sheet_name="Svenska som andraspråk")
df_list = [df_en,df_se,df_math,df_sfi] # Add the frames to a list in case i need to do something to them all
title_list = ["Engelska","Svenska","Matematik","Svenska som andraspråk"]

```

## Data Cleaning

##### Looping through list and fixes the collumnnames

```python

# Looping through to change column names
for i in range(len(df_list)):   
    df_list[i].columns = [
        "Plats",
        "Huvudman",
        "Totalt(A-F)",
        "Flickor(A-F)",
        "Pojkar(A-F)",
        "Totalt(A-E)",
        "Flickor(A-E)",
        "Pojkar(A-E)",
        "Totalt(poäng)",
        "Flickor(poäng)",
        "Pojkar(poäng)",
    ]
    

    #print(dataframe.info()) # printing info for testing purposes
print(df_sfi) # Realise SFI has dirty data dont see any real way to clean it. rebating wether to remove or set to 0
df_sfi.iloc[3,2:] = 0 #Decided to set to  0

    
```

##### PlotBuilding

```python
fig,axs = plt.subplots(nrows = 1,ncols = 4, figsize= (len(df_list)*7,5)) #Creates subplots that scale with the length of the list
color_list = ["red","blue","green","orange"]
for i in range(len(df_list)):
    
    axs[i].set_ylim(0,20)
    axs[i].set_xlabel("Huvudmän",fontsize = 10,fontweight = "bold")    # Talked to ludvig and he told me how to change font when setting labels
    axs[i].set_ylabel("Poäng (tot)",fontsize = 10,fontweight = "bold")
    axs[i].set_title(title_list[i],fontsize = 20,fontweight = "bold")
    axs[i].color = color_list
    axs[i].bar(df_list[i]["Huvudman"],df_list[i]["Totalt(poäng)"],color= color_list)
fig.savefig("visualiseringar/points_per_huvudman.png")      #Asked chatgpt how to save a fig to png and learned the savefig() method


```
